###### Model bulding

```python
import statsmodels.formula.api as smf
```

```python
model1 = smf.ols('Experience~salary',data = df).fit()
```

###### Model Testing

```python
# findingcoefficient parameters
model1.params
```

```python
# finding tvalues and pvalues
model1.tvalues, model1.pvalues
```

```python
model1.summary()
```

```python
# finding rsquarded values
model1.rsquared , model1.rsquared_adj
```

```python
model2 = smf.ols('np.log (Experience)~np.log(salary)',data = df).fit()
```

```python
# findingcoefficient parameters
model2.params
```

```python
# finding tvalues and pvalues
model2.tvalues, model2.pvalues
```

```python
model2.summary()
```

```python
# finding rsquarded values
model2.rsquared , model2.rsquared_adj
```

Model Prediction

```python
# automatic Prediction for say sorting time 6,8
new_df = pd.Series([6,8])
new_df
```

```python
data_pred = pd.DataFrame(new_df,columns = ['salary'])
data_pred
```

```python
model1.predict(data_pred)
```
